```python
import os

# ëŒ€ìƒ í´ë” ê²½ë¡œ ì„¤ì •(ì‹¤ì œ ë°±ì—”ë“œ ê²½ë¡œë¡œ ë³€ê²½ í•„ìš”)
TARGET_FOLDER_PATH = r"C:\Pyg\REACT\6_todolist"

# ê²°ê³¼ë¬¼ì„ ì €ì¥í•  íŒŒì¼ëª…
OUTPUT_FILENAME = "project_full_context.txt"

# ìˆ˜ì§‘í•  íŒŒì¼ í™•ì¥ì 
TARGET_EXTENSIONS = ['.mjs', '.js', '.ts', '.py', '.java', '.go', '.json', '.yaml', '.yml', '.sh', '.rb', '.php', '.html', '.css', '.scss', '.md', '.jsx', '.tsx']

# ì œì™¸í•  í´ë”
IGNORE_DIRS = {
    'node_modules', 'venv', '.git', '__pycache__', 
    'dist', 'build', '.idea', '.vscode', 'coverage',
}

# ì œì™¸í•  íŒŒì¼ (íŒŒì¼ ì´ë¦„ë§Œ ì •í™•íˆ)
IGNORE_FILES = {
    'package-lock.json',
    'yarn.lock',
    '.DS_Store'
}

print("âœ… ì„¤ì •ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
print(f"   - ëŒ€ìƒ í´ë”: {TARGET_FOLDER_PATH}")
print(f"   - ìˆ˜ì§‘ í™•ì¥ì: {TARGET_EXTENSIONS}")
```

```python
def collect_backend_code(root_dir, output_file):
    if not os.path.exists(root_dir):
        print(f"âŒ ì˜¤ë¥˜: ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤ -> {root_dir}")
        return

    print(f"ğŸ“‚ í´ë” ìŠ¤ìº” ì‹œì‘: {root_dir}")
    file_count = 0
    
    try:
        with open(output_file, 'w', encoding='utf-8') as f_out:
            f_out.write(f"# Project Source Code Integration\n")
            f_out.write(f"# Target Directory: {root_dir}\n")
            f_out.write("="*60 + "\n\n")

            for root, dirs, files in os.walk(root_dir):
                # ì œì™¸í•  í´ë” ê±´ë„ˆë›°ê¸°
                dirs[:] = [d for d in dirs if d not in IGNORE_DIRS]

                for file in files:
                    # ì—¬ê¸°ì„œ ë¨¼ì € íŒŒì¼ ì´ë¦„ì„ ê²€ì‚¬
                    if file in IGNORE_FILES:
                        continue

                    _, ext = os.path.splitext(file)
                    
                    if ext in TARGET_EXTENSIONS:
                        file_path = os.path.join(root, file)
                        try:
                            with open(file_path, 'r', encoding='utf-8') as f_in:
                                content = f_in.read()
                                f_out.write(f"### FILE: {file_path}\n")
                                f_out.write("-" * 30 + "\n")
                                f_out.write(content)
                                f_out.write(f"\n\n{'='*60}\n\n")
                                file_count += 1
                                print(f"  -> ìˆ˜ì§‘ë¨: {file}")
                        except Exception as e:
                            print(f"âš ï¸ ì½ê¸° ì‹¤íŒ¨: {file_path}")
                            
        print(f"\n ì‘ì—… ì™„ë£Œ! (ì œì™¸ íŒŒì¼ì€ ê±´ë„ˆë›°ì—ˆìŠµë‹ˆë‹¤)")
        print(f"   - ì´ {file_count}ê°œì˜ íŒŒì¼ì´ í†µí•©ë˜ì—ˆìŠµë‹ˆë‹¤.")
        
    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
```

```python
# ì‹¤í–‰
if __name__ == "__main__":
    collect_backend_code(TARGET_FOLDER_PATH, OUTPUT_FILENAME)
```
